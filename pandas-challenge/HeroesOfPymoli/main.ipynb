```python
#Importing Modules

import pandas as pd
import numpy as nm
```

```python
#Using Pandas to read/open the data file

df=pd.read_csv("Resources/purchase_data.csv")
df.head(10)
```

### Player Count

```python
# Total Number of Players
unique_players=df["SN"].unique()
total_players=len(unique_players)

player_count_df=pd.DataFrame({"Total Players": [total_players]})

player_count_df
```

### Purchasing Analysis (Total)

```python
# Number of Unique Items
unique_items= df["Item ID"].unique()
total_count=len(unique_items)

#Average Purchase Price
avg_df=df["Price"].mean() 

# Total Number of Purchases
Total_Purchases= df["Purchase ID"].count()
Total_Purchases

# Total Revenue
sum_df=df["Price"].sum()
```

```python
summary_table=pd.DataFrame({"Number of Unique Items": total_count,
                            "Average Price": [avg_df],
                            "Number of Pruchases": [Total_Purchases],
                            "Total Revenue": [sum_df]})

summary_table=summary_table.round(2)

summary_table["Average Price"]=summary_table["Average Price"].map("${:,.2f}".format)
summary_table["Total Revenue"]=summary_table["Total Revenue"].map("${:,.2f}".format)
summary_table
```

### Purchasing Analysis (Gender)

```python
#Creating a df just for Gender and unique Id's
player_demographics = df.loc[:, ['Gender', 'SN', "Age"]]
player_demographics = player_demographics.drop_duplicates()
```

```python
#Purchase Count by Gender
gender_count = df.groupby(["Gender"]).count()["Price"]

#Average Purchase Price
gender_avg_price = df.groupby(["Gender"]).mean()["Price"]

#Total Purchases by Gender
gender_total_purchase = df.groupby(["Gender"]).sum()["Price"]

#Average Total Purchase per person by Gender
avg_price_gender=gender_total_purchase/count
```

```python
gender_table=pd.DataFrame({"Purchase Count": gender_count,
                            "Average Purchase Price": gender_avg_price,
                            "Total Purchase Value" : gender_total_purchase,
                            "Average Purchase Total per Person by Gender" : avg_price_gender})

gender_table=gender_table.round(2)
gender_table["Average Purchase Price"]=gender_table["Average Purchase Price"].map("${:,.2f}".format)
gender_table["Total Purchase Value"]=gender_table["Total Purchase Value"].map("${:,.2f}".format)
gender_table["Average Purchase Total per Person by Gender"]=gender_table[ "Average Purchase Total per Person by Gender"].map("${:,.2f}".format)
gender_table
```

### Age Demographics

```python
#Bins
bins=[0,9,14,19,24,29,34,39,100]

labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

player_demographics["Purchase Count Bins"]=pd.cut(player_demographics["Age"],bins,labels=labels)

age_value=player_demographics["Purchase Count Bins"].value_counts()

percent_age=((age_value/total_players)*100)
```

```python
#Final DF for Age Demographics

age_table=pd.DataFrame({"Total Count": age_value ,
                            "Percentage of Players": percent_age})

age_table=age_table.sort_index()
age_table["Percentage of Players"]=age_table["Percentage of Players"].map("{:,.2f}%".format)
age_table
```

### Purchase Analysis (Age)

```python
#Made a new column in my dataframe for Age Ranges
df["Age Ranges"]=pd.cut(df["Age"],bins,labels=labels)

#run aggregate calculations for each bin
bin_count = df.groupby(["Age Ranges"]).count()["Price"]

#Calculate Average Purchase Total per person for each bin
avg_bin=df.groupby(["Age Ranges"]).mean()["Price"]

#Total Purchase Value
total_bin_purchase=df.groupby(["Age Ranges"]).sum()["Price"]

#Avg Total Purchase per Person
total_bin_avg= total_bin_purchase/age_value

#Convert to dataframe
purchase_df=pd.DataFrame({"Purchase Count": bin_count,
                            "Average Purchase Price": avg_bin,
                         "Total Purchase Value": total_bin_purchase,
                         "Avg Total Purchase per Person": total_bin_avg})
#Formatting & Display
purchase_df["Average Purchase Price"]=purchase_df["Average Purchase Price"].map("${:,.2f}".format)
purchase_df["Total Purchase Value"]=purchase_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_df["Avg Total Purchase per Person"]=purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_df
```

### Top Spenders

```python
#Create data frame with SN and Price Column

SNP_df=df.loc[:,["SN","Price"]]

#Purchase Count
SNP_count = df.groupby(["SN"]).count()["Price"]

#Average Purchase Price
SNP_avg = df.groupby(["SN"]).mean()["Price"]


#Total Purchase Value

SN_value = df.groupby(["SN"]).sum()["Price"]

SN_value.sort_values(ascending=False)

#Make a dataframe

Top_Spenders=pd.DataFrame({"Purchase Count": SNP_count,
                           "Average Purchase Price": SNP_avg,
                          "Total Purchase Value":SN_value})

Top_Spenders=Top_Spenders.sort_values("Total Purchase Value", ascending=False)

#Formatting
Top_Spenders["Average Purchase Price"]=Top_Spenders["Average Purchase Price"].map("${:,.2f}".format)
Top_Spenders["Total Purchase Value"]=Top_Spenders["Total Purchase Value"].map("${:,.2f}".format)

Top_Spenders.head(5)
```

### Most Popular Items

```python
#Identify the 5 most popular items by purchase count, then list (in a table):
item_df=df.loc[:, ["Item ID", "Item Name", "Price"]]

item_df.head()

#Purchase Count
item_df = df.groupby(["Item ID", "Item Name"]).count()["Price"]

#Item Price
price_df=df.groupby(["Item ID", "Item Name"]).mean()["Price"]

#Total Purchase Value
total_df=df.groupby(["Item ID", "Item Name"]).sum()["Price"]

#Make a dataframe 

item_data=pd.DataFrame({"Purchase Count": item_df,
                        "Item Price": price_df,
                         "Total Purchase Value":total_df})

Popular_df=item_data.sort_values("Purchase Count", ascending=False)

#Format
Popular_df["Item Price"]=Popular_df["Item Price"].map("${:,.2f}".format)
Popular_df["Total Purchase Value"]=Popular_df["Total Purchase Value"].map("${:,.2f}".format)


Popular_df.head()
```

### Most Profitable Items

```python
most_profitable=item_data.sort_values("Total Purchase Value", ascending=False)


#Format
most_profitable["Item Price"]=most_profitable["Item Price"].map("${:,.2f}".format)
most_profitable["Total Purchase Value"]=most_profitable["Total Purchase Value"].map("${:,.2f}".format)

most_profitable.head(5)
```
